```python
import json

file_path = "final_traced_dataset.jsonl"  

with open(file_path, 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        if i >= 20:  
            break
            
        data = json.loads(line)
        
        input_var = data.get("input")
        output_var = data.get("answer") 
        code_snippet = data.get("code")
        traced_info = data.get("traced_info")
        
        print(f"\n{'='*80}")
        print(f"  ENTRY {i+1}")
        print(f"{'='*80}")
        
        print(f"\n[INPUT]:\n{input_var}")
        print(f"\n[EXPECTED OUTPUT]:\n{output_var}")
        print(f"\n[CODE]:\n{code_snippet}")
        
        print("\n[TRACED INFO]:")
        print(json.dumps(traced_info, indent=4, sort_keys=False))
        
        print("\n" + "-"*80)
```
